## Первая программа. Встроенная функция `print()`

Перед вами уже знакомая ячейка кода, которую вы можете запустить.

Но, так как она пустая, ничего не произойдет. Чтобы научиться писать код, изучите встроенную функцию языка Python — `print()`.

Встроенная функция — термин в словарном запасе Python, зарезервированный для выполнения определенного действия, которое заранее установлено разработчиками (сопровождается скобками).

Функция в программировании — это маленькая программа, которая выполняет какую-то операцию. Функцию можно вызвать. После имени функции следуют круглые скобки, в которых указывают аргументы (над каким объектом будет выполнена операция) и параметры (с какими условиями).

Функция `print()` выводит на экран данные, которые вы передаете внутрь скобок. При вызове функции в Python скобки обязательны. Иначе компьютер не поймет, что вы вызываете именно функцию.

### Спец символы:
```
\n: Новая строка
\t: Горизонтальная табуляция
\': Одинарная кавычка
\": Двойная кавычка
\\: Экранирование
```

1. `\n`: Это символ новой строки. Когда он встречается в строке, текст, следующий за ним, начнется с новой строки. Это удобно для форматирования текста или вывода информации в читаемом виде.

2. `\t`: Это символ горизонтальной табуляции. Он используется для создания отступа, как если бы вы нажали клавишу табуляции. Это помогает структурировать данные, делая их более удобными для восприятия.

3. `\'`: Это экранирование одинарной кавычки. Если тебе нужно использовать одинарную кавычку внутри строки, она должна быть предварена обратной косой чертой, чтобы интерпретатор понял, что это часть текста, а не конец строки.

4. `\"`: Точно так же, это экранирование для двойной кавычки. Если ты хочешь включить двойные кавычки в строку, используй их вместе с обратной косой чертой.

5. `\\`: Это просто символ обратной косой черты. Иногда нужно вставить ее в строку, и для этого используется двойная обратная косая черта.


````python
# \n: Новая строка
print("Первая строка\nВторая строка")
````

````python
# \t: Горизонтальная табуляция
print("Столбец 1\tСтолбец 2\tСтолбец 3")
````

````python
# \': Одинарная кавычка
print('Это \'одинарная\' кавычка')
````

````python
# \": Двойная кавычка
print("Это \"двойная\" кавычка")
````

````python
# \\: Обратная косая черта
print("Это обратная косая черта: \\")
````

## **Арифметические операторы**


Язык **Python** позволяет давать компьютеру задания по вычислениям. В арсенале вы имеете следующие операции:
1.	Сложение: `+`
2.	Вычитание: `-`
3.	Умножение: `*`
4.	Деление: `/`
5.	Возведение в степень: `**`
6.	Целочисленное деление: `//`
7.	Остаток от деления: `%`

Остановимся подробнее на последних двух.

**Целочисленное деление** возвращает только целую часть операции деления. Например, при делении 10 // 3 в ответе вы получите 3. Потому что в числе 10 помещается только три тройки. Остаток от деления (1) отбрасывается.

**Операция взятия остатка** возвращает остаток от деления нацело. То есть при таком же делении 10 на 3 в остатке вы получите 1. Так как в числе 10 помещается три тройки, а в остатке остается 1. Операция записывается так: 10 % 3




del.jpg

А теперь комбинируем арифметические операции с функцией `print()`. Напишем программу, которая будет производить все изученные операции с цифрами 10 и 3, а ответ выводить на экран:

````python
print(10 + 3)                             # Сложение
print(10 - 3)                             # Вычитание
print(10 * 3)                             # Умножение
print(10 / 3)                             # Деление
print(10 ** 3)                            # Возведение в степень
print(10 // 3)                            # Целочисленное деление (10 = 3 + 3 + 3 + 1 = 3 * 3 + 1)
print(10 % 3)                             # Остаток от деления (10 = 3 + 3 + 3 + 1 = 3 * 3 + 1)
````

Если взглянуть на выводимые данные, не совсем понятно, где результат деления, где результат сложения и так далее. Попробуйте самостоятельно написать в новой ячейке код, который будет выводить на экран такой результат:

        10 + 3 = 13
        10 - 3 = 7
        10 * 3 = 30
        10 / 3 = 3.333
        10 ** 3 = 1000
        10 // 3 = 3
        10 % 3 = 1

## `f-строка`

**F-строка** — это способ форматирования строк в Python, введенный в версии 3.6 и более поздних. Она позволяет включать значения переменных и выражений непосредственно в строки, делая код более читаемым и удобным. Давай рассмотрим примеры:

1. **Вставка переменных:**

````python
name = "Иван"
age = 30
greeting = f"Привет, {name}! Тебе уже {age} лет."
print(greeting)
````

   В этом примере значения переменных `name` и `age` вставляются непосредственно в строку `greeting` с использованием f-строки.

2. **Вычисление выражений:**

````python
x = 5
y = 10
result = f"Сумма чисел {x} и {y} равна {x + y}."
print(result)
````

   Здесь f-строка используется для вставки результата выражения `{x + y}` в строку.

3. **Форматирование чисел:**

````python
pi = 3.141592653589793
formatted_pi = f"Значение числа π с округлением: {pi:.2f}"
print(formatted_pi)
````

 В этом примере `:.2f` указывает на формат числа с двумя знаками после запятой.

## **Типы данных**

**Тип данных** – некоторое представление данных в программе (числа, строки, списки и т. д.). Язык **Python** поддерживает множество типов данных, вот самые основные:

* **int** *(integer numbers)* – целые числа;
* **float** *(floating-point numbers)* – числа с плавающей точкой (действительные числа);
* **str** *(string)* – строки;
* **bool** *(boolean)* – логический тип.


````python
integer = 3                               # Переменная типа int
float_number = 1.48376                    # Переменная типа float
text = "Hello, World!"                    # Строковая переменная (тип str)
isTrue = True                             # Логическая переменная (тип boolean),
                                          # может принимать только два значения
                                          # True или False (истина / ложь)
````

Функция `type()` определяет тип значения своего аргумента.

````python
print(type(integer))                      # Вывод типа значения переменных
print(type(float_number))
print(type(text))
print(type(isTrue))
````

Также можно преобразовать значение одного типа к другому типу данных, используя следующие функции:
* `int()` –  возвращает целое значение аргумента;
* `float()` – возвращает нецелочисленное (действительное) значение аргумента;
* `bool()` – возвращает булево (логическое) значение аргумента;
* `str()`– возвращает строковое значение аргумента.

Например:


````python
number = 1050
print(number, type(number))               # Проверка типа значения переменной
print(number * 3)
````

````python
float_num = float(number)                 # Преобразование значения переменной к другому типу
print(float_num, type(float_num))
print(float_num / 8)
````

С помощью `float(number)` вы преобразовали значение переменной `number` к нецелочисленному типу `float`.

````python
text = str(number)
print(text, type(text))
print(text * 4)                           # Умножение строки на целое число - это повторение
                                          # строки несколько раз (конкатенация)
````

С помощью `str(number)` вы преобразовали значение переменной `number` к строковому типу `str`.

Обратите внимание на строчку кода: `print(text * 4)`.

В данном случае вы выполнили операцию конкатенации (объединения) строк. Поскольку тип переменной `text` - это строка, то **Python** не сможет осуществить операцию умножения (так как эта операция применима только к числовым типам данных). Вместо этого **Python** повторит строку 4 раза или другими словами: выполнит конкатенацию строки `text` 4 раза.

##**Ввод информации** `input()`

Для ввода информации с клавиатуры предусмотрена стандартная функция `input()`.

Она используется таким образом:
1. Запустите ячейку ниже;
2. Введите произвольное сообщение в появившееся поле ввода;
3. Нажмите клавишу **Enter**.

````python
input()
````

В качестве аргумента функции можно указать **строку-приглашение**:

````python
input('Введите что-нибудь, пожалуйста: ') # Использование строки-приглашения к вводу
````

Введенную пользователем информацию можно сохранить в переменную:

````python
some_value = input('Введите любое число, пожалуйста: ')  # Запись введенной информации
                                                        # в переменную, на выходе тип str (строка)
print(type(some_value))
````

<h4><hr>
<strong>Важно: любая введеная с помощью функции input() информация имеет строковый тип данных (str).</strong>

````python
print(int(some_value) * 10)             # Преобразование значения переменной к числовому типу (int)
````

## **Логический тип - `bool` (boolean)**

Логический тип (`bool`) — тип данных с двумя возможными значениями: `True` (истина) и `False` (ложь). Он нужен для работы с выражениями, их сравнения, проверки истинности условий.

Как вы знаете, логический тип данных может принимать только два значения: `True` или `False`. Это два ключевых слова, зарезервированных в Python так же, как и названия встроенных функций, будь то `print` или `input`. Только для использования этих ключевых слов не нужно дописывать к ним скобки, как вы делали в случае с функциями.

Очень часто этот тип данных используют для обозначения логической последовательности в программе. То есть определения истинности выражения:

````python
print(True, type(True))
print(False, type(False))
````

Простое логическое выражение имеет вид:

    <арифметическое выражение> <знак сравнения> <арифметическое выражение>

Или:

    <строка> <знак сравнения> <строка>

Или в более общем виде:

    <данные> <знак сравнения> <данные>.

Результат логического выражения — всегда значение типа данных `boolean`.

Сравните два числа:

````python
print(1 > 2)  # <данные> <знак сравнения> <данные>
print(1 < 2)  # <данные> <знак сравнения> <данные>
````

`1 > 2`

На что Python дает ответ: `False`.


## **Операции сравнения**

Это операции, существующие в Python для сравнения значений, объектов или арифметических выражений:

* `==` – равно;
* `!=` – не равно;
* `<` – меньше;
* `<=` – меньше или равно;
* `>` – больше;
* `>=` – больше или равно.


Когда и как работают различные операторы сравнения – тема отдельной лекции. Важно то, что для каждого типа данных каждая операция сравнения определена по-своему. Например, операция сравнения `<` для двух чисел – это не то же самое, что `<` для двух строк.

>**Важно:** не путайте операцию присваивания значения переменной, обозначаемую в языке Python одиночным знаком «равно», и операцию сравнения (два знака «равно»). Присваивание и сравнение — разные операции.


````python
# <арифметическое выражение> <знак сравнения> <арифметическое выражение>
print(2 + 2 == 2 * 2)
````

````python
# <арифметическое выражение> <знак сравнения> <арифметическое выражение>
print(3 + 3 == 3 * 3)
````

````python
# <арифметическое выражение> <знак сравнения> <арифметическое выражение>
print(3 + 3 != 3 * 3)
````

## **Логические операторы `and`, `or` и `not`**


>**Для записи сложных логических выражений часто пользуются логическими операторами «и» (`and`), «или» (`or`) и «не» (`not`)**


* `and` возвращает True, если оба выражения равны **True**;
* `or` возвращает True, если хотя бы одно из двух выражений равно **True**;
* `not` меняет значение выражения на противоположное.

image.png

Выражение с оператором «и» (`and`) проверяет истинность левой и правой частей от себя, и если обе части истинны, то возвращает `True`:

````python
print(a > 20 and b > 40)
````

````python
print(a > 20 and b < 40)
````

Операции, где происходит более одного сравнения, называют **составными**.

Так как в `a` и `b` лежит число 36, то конструкция `a > 20` возвращает `True`, `b > 40` возвращает `False`. А `False and True` – это `False`.

Выражение с оператором «или» (`or`) проверяет истинность левой и правой частей. Если хотя бы одна из них истинна, то даст `True`:

````python
print(a > 20 or b > 40)
````

````python
print(a > 20 or b < 40)
````

````python
print(a == 20 or b == 40)
````

Выражение с оператором «не» (`not`) проверяет истинность правой части от себя, и возвращает противоположное ему значение:

````python
print(not a < 20)
print(not a > 20)
````

Составим более сложное логическое выражение:

````python
x = 5
y = 10

result = not(((x < y) or (x >= y)) == (x != y))  # Вычисление сложного логического выражения
print(result)
````

        result = not(((x < y) or (x >= y)) == (x != y))
Разберем выражение выше по частям. Сначала выполняются операции в скобках. `or` и `and` выполняются, когда определены их аргументы:
1. `x < y` – x меньше y? **Да (True)**
2. `x >= y` – x больше либо равен y? **Нет (False)**
3. `(x < y) or (x >= y)` - подставляя вместо вычисленных подвыражений **True** и **False**, получим результат `True or False` = **Да (True)**
4. `x != y` – x не равен y? **Да (True)**
5. `((x < y) or (x >= y)) == (x != y)` – результаты значений в скобках уже подсчитаны. **True** равно **True**? **Да (True)**
6. в самом конце оператор `not` меняет значение на противоположное – исходное значение равно **True**, значит на выходе - **False**.
